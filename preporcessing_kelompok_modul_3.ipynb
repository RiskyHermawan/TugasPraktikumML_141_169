#Preprocessing

```python
#Gather data training dengan ukuran gambar 250 x 250

import os
import cv2
import numpy as np

# Gather data train inisialisasi kosongan
train_data = []
train_label = []

train_dir=os.path.join('/content/Splitingflowers/training')
for r, d, f in os.walk(train_dir):
    for file in f:
        if ".jpg" in file:
            imagePath = os.path.join(r, file)
            image = cv2.imread(imagePath)
            image = cv2.resize(image, (250,250))
            train_data.append(image)
            label = imagePath.split(os.path.sep)[-2]
            train_label.append(label)

train_data = np.array(train_data)
train_label = np.array(train_label)
```

```python
#Gather data validation dengan ukuran gambar 250 x 250

val_data = []
val_label = []

vali_dir=os.path.join('/content/Splitingflowers/validation')
for r, d, f in os.walk(vali_dir):
    for file in f:
        if ".jpg" in file:
            imagePath = os.path.join(r, file)
            image = cv2.imread(imagePath)
            image = cv2.resize(image, (250,250))
            val_data.append(image)
            label = imagePath.split(os.path.sep)[-2]
            val_label.append(label)

val_data = np.array(val_data)
val_label = np.array(val_label)
```

```python
#Gather data test dengan ukuran gambar 250 x 250

test_data = []
test_label = []

test_dir=os.path.join('/content/Splitingflowers/test')
for r, d, f in os.walk(test_dir):
    for file in f:
        if ".jpg" in file:
            imagePath = os.path.join(r, file)
            image = cv2.imread(imagePath)
            image = cv2.resize(image, (250,250))
            test_data.append(image)
            label = imagePath.split(os.path.sep)[-2]
            test_label.append(label)

test_data = np.array(test_data)
test_label = np.array(test_label)
```

```python
#Tampilkan shape dari data train, data validation dan data test

print("Train Data = ", train_data.shape)
print("Train Label = ", train_label.shape)
print("Validation Data = ", val_data.shape)
print("Validation Label = ", val_label.shape)
print("Test Data = ", test_data.shape)
print("Test Label = ", test_label.shape)
```

```python
# Normalisasi dataset

print("Data sebelum di-normalisasi ", train_data[0][0][0])
x_train = train_data.astype('float32') / 255.0
x_val = val_data.astype('float32') / 255.0
x_test = test_data.astype('float32') / 255.0
print("Data setelah di-normalisasi ", x_train[0][0][0])
```

```python
# membuat dan menampilkan hasil dari label encoder

from sklearn.preprocessing import LabelEncoder
print("Label sebelum di-encoder ", train_label[0:10])

lb = LabelEncoder()
y_train = lb.fit_transform(train_label)
y_val = lb.fit_transform(val_label)
y_test = lb.fit_transform(test_label)

print("Label setelah di-encoder ", y_train[0:10])
```

```python
np.save("x_train", x_train)
np.save("x_val", x_val)
np.save("x_test", x_test)
np.save('y_train', y_train)
np.save('y_val', y_val)
np.save('y_test', y_test)
```
